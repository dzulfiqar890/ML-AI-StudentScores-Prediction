```python
df = dataset.copy()
```

```python
print("Data Sebelum Pemeriksaan Duplikat:")
print(df.shape)
```

```python
duplicate_rows_before = df[df.duplicated()]
duplicate_rows_before
```

```python
df = df.drop_duplicates()
```

```python
print("\nData Setelah Pemeriksaan Duplikat:")
print(df.shape)
```

```python
df.isna().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Outlier Analysis
sns.boxplot(x="Hours", data=df)

# Menambahkan label sumbu dan judul
plt.xlabel("Hours")
plt.ylabel("Frequency")  # Anda bisa menggunakan "Years of Experience" atau "Tahun Pengalaman"
plt.title("Boxplot Hours of Student Scores")

# Menampilkan plot
plt.show()
```

```python
X = df['Hours']
X.head()
```

```python
y = df['Scores']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.75,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np

# Jika masih dalam bentuk DataFrame, konversi ke NumPy array
if hasattr(X_train, "to_numpy"):
    X_train = X_train.to_numpy()
if hasattr(X_test, "to_numpy"):
    X_test = X_test.to_numpy()

# Tambahkan axis baru jika X_train masih 1D
if X_train.ndim == 1:
    X_train = X_train[:, np.newaxis]
if X_test.ndim == 1:
    X_test = X_test[:, np.newaxis]
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train,y_train)
```

```python
y_pred_train = lr_model.predict(X_train)
y_pred_train
```

```python
y_train
```

```python
y_pred = lr_model.predict(X_test)
```

```python
X_test
```

```python
y_pred
```

```python
c = [i for i in range (1,len(y_test)+1,1)]
plt.plot(c,y_test,color='r',linestyle='-',label='Actual Data')
plt.plot(c,y_pred,color='b',linestyle='dashed',label='Prediction')
plt.xlabel('Scores')
plt.ylabel('index')
plt.title('Prediction with Linear Regression')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
rsq = r2_score(y_test,y_pred)
```

```python
print('r square Linear Regression:',rsq)
```

```python
print('Intercept of the Linear Regression model:',lr_model.intercept_)
print('Coefficient of the line Linear Regression:',lr_model.coef_)
```
